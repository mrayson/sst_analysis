```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import animation
import glob
import dask

import zarr
```

```python
sites = {
    'NIN100':{'y':-21.84986667,'x':113.9064667},
    'ITFTIS':{'x': 127.5577, 'y':-9.819217}   ,
    'KIM200':{'x':121.243217 , 'y':-15.534517} ,  
    'KIM400':{'x': 121.114967, 'y':-15.22125}   ,
    'PIL200':{'x': 115.9154, 'y':-19.435333}   ,
    'NWSBAR':{'y':-20.76128333,'x':114.7586167},
    'NWSROW':{'y':-17.75801667,'x':119.9061},
    'NWSBRW':{'y':-14.23543333,'x':123.1623833},
    'NWSLYN':{'y':-9.939416667,'x':130.3490833},
    'BB250':{'x':123.34613 , 'y':-13.75897}   ,
}
x = np.array([sites[nn]['x'] for nn in sites.keys()])
y = np.array([sites[nn]['y'] for nn in sites.keys()])
```

```python
outpath = '/home/mrayson/scratch/HIMAWARI/raw_zarr'
ds1 = xr.open_zarr(outpath).sel(lon=slice(x.min()-0.2,x.max()+0.2),
                               lat=slice(y.max()+0.2, y.min()+0.2))
ds1 = ds1.chunk({'time':1000,'lat':10,'lon':10})
ds1
```

```python
outpath = '/home/mrayson/scratch/HIMAWARI/raw_zarr_v2'
ds2 = xr.open_zarr(outpath).sel(lon=slice(x.min()-0.2,x.max()+0.2),
                               lat=slice(y.max()+0.2, y.min()+0.2))
ds2 = ds2.chunk({'time':1000,'lat':10,'lon':10})
ds2
```

```python
ds1 = ds1.persist()
ds2 = ds2.persist()
```

```python
delayed_objs = []
for nn in sites.keys():
    x,y = sites[nn]['x'], sites[nn]['y']
    myds1 = ds1.sel(lon=x, lat=y, method='nearest')#.chunk({'time':-1})
    myds2 = ds2.sel(lon=x, lat=y, method='nearest')#.chunk({'time':-1})
    
    print('Loading site {}...'.format(nn))
    myds = xr.concat([myds1, myds2], dim='time').compute()

    outfile = '/scratch/pawsey0106/mrayson/HIMAWARI/JAXA_Himawari_SST_IMOS_Site_{}_201601_202106.nc'.format(nn)
    #with dask.config.set(scheduler='threads'):
    myds.to_netcdf(outfile)
    print(outfile)
```
